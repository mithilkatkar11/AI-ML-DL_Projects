```python
import pandas as pd
```

```python
weather_data = pd.read_csv("weather_data1.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=weather_data, x="Days", y="Temperature_Avg", label='Temperature')
sns.lineplot(data=weather_data, x="Days", y="Humidity_AVg(%)", label='Humidity')
plt.xlabel("Days")
plt.ylabel("Value")
plt.title('Temperature and Humidity Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(weather_data['Days'],weather_data['AQI Index'],label = 'AQI Index')
plt.title("AQI Index Over Time")
plt.xlabel("Days")
plt.ylabel("AQI Index")
plt.legend()
plt.grid(True)
plt.show()
```

```python
fig,(ax1, ax2) = plt.subplots(1,2, figsize = (10,4))
ax1.plot(weather_data['Days'],weather_data['Wind_Speed_Avgmph)'],label = 'Wind Speed')
ax1.set_xlabel("Days")
ax1.set_ylabel("Wind Speed")
ax1.set_title("Wind Speed over Time")
ax1.legend()
ax1.grid(True)

ax2.plot(weather_data['Days'],weather_data['Precipitation (in)'],label = 'Precipitation', c = 'y')
ax2.set_xlabel("Days")
ax2.set_ylabel("Precipitation")
ax2.set_title("Precipitation over Time")
ax2.legend()
ax2.grid(True)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=weather_data, x="Temperature_Avg", y="AQI Index", hue="Precipitation (in)",palette='deep')
plt.xlabel("Temperature")
plt.ylabel("AQI Index")
plt.title('Scatter Plot of Temperature vs. AQI Index')
plt.show()
```

```python
fig,(ax1, ax2) = plt.subplots(1,2, figsize = (10,4))
ax1.scatter(weather_data['Days'],weather_data['AQI Index'], c = 'y', label = 'AQI Index')
ax1.set_xlabel("Days")
ax1.set_ylabel("AQI Index")
ax1.set_title("Scatter Plot of AQI Index")
ax1.legend()

ax2.scatter(weather_data['Days'],weather_data['Precipitation (in)'],c = 'b',label = 'Precipitation')
ax2.set_title("Scatter Plot of Precipitation")
ax2.set_xlabel("Days")
ax2.set_ylabel("Precipitation")
plt.legend()
```

```python
fig,(ax1, ax2) = plt.subplots(1,2, figsize = (10,4))
ax1.scatter(weather_data['Days'],weather_data['Temperature_Avg'], c = 'r', label = 'Temperature')
ax1.set_xlabel("Days")
ax1.set_ylabel("Temperature")
ax1.set_title("Scatter Plot of Temperature")
ax1.legend()

ax2.scatter(weather_data['Days'],weather_data['Humidity_AVg(%)'],c = 'b',label = 'Humidity')
ax2.set_title("Scatter Plot of Humidity")
ax2.set_xlabel("Days")
ax2.set_ylabel("Humidity")
plt.legend()
```

```python
fig,(ax1, ax2) = plt.subplots(1,2, figsize= (10,4)) 
ax1.scatter(weather_data['Days'], weather_data['Wind_Speed_Avgmph)'], c = 'g' , label = 'Wind Speed')
ax1.set_xlabel("Days")
ax1.set_ylabel("Wind Speed")
ax1.set_title("Scatter Plot of Wind Speed")
ax1.legend()

ax2.scatter(weather_data['Days'], weather_data['Temperature_Avg'], c = 'r', label = 'Temperature')
ax2.set_xlabel("Days")
ax2.set_ylabel("Temperature")
ax2.set_title("Scatter Plot of Temperature")
ax2.legend()
```

```python
correlation_matrix = weather_data.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heat Map')
plt.show()
```

```python
from ipywidgets import interact
```

```python
weather_data.columns
```

```python
variables = ['Temperature_Avg','Humidity_AVg(%)','Wind_Speed_Avgmph)','Precipitation (in)','AQI Index']
```

```python
def interact_plot(variables):
    plt.figure(figsize=(10,6))
    plt.plot(weather_data['Days'],weather_data[variables])
    plt.title(f"{variables} Over Time")
    plt.xlabel("Days")
    plt.ylabel(f"{variables}")
    plt.grid(True)
    plt.show()
```

```python
interact(interact_plot, variables = ['Temperature_Avg','Humidity_AVg(%)','Wind_Speed_Avgmph)','Precipitation (in)','AQI Index'])
```
