## Part 1 : Exploring Data Visualization in Python with Matplotlib


```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df["species"] = iris.target
```

```python
iris_df
```

```python
iris_df.isnull().sum()
```

```python
features = iris.feature_names
```

```python
features
```

```python
iris.target_names
```

```python
for feature in features:
    for species_id, species_name in enumerate(iris.target_names):
        subset = iris_df[iris_df["species"] == species_id]
        plt.plot(subset[feature], label=species_name)  
    plt.title(f"{feature} vs. Species")
    plt.xlabel("Index")
    plt.ylabel("Value")
    plt.legend(title="Species")
    plt.show()
```

```python
sns.set(style="whitegrid")
sns.pairplot(iris_df, hue="species", palette="Set1")
plt.show()
```

```python
for i in range(len(features) - 1):
    for j in range(i + 1, len(features)):
        feature_x = features[i]
        feature_y = features[j]        

        plt.figure(figsize=(8, 6))
        sns.scatterplot(data=iris_df, x=feature_x, y=feature_y, hue="species", palette="Set1")
        plt.title(f"{feature_x} vs {feature_y} by Species")
        plt.xlabel(feature_x)
        plt.ylabel(feature_y)
        plt.legend(title="Species")        
        plt.show()
```

```python
for feature in features:  
    for species_id, species_name in enumerate(iris.target_names):
        subset = iris_df[iris_df["species"] == species_id]
        sns.histplot(subset[feature],bins =20, kde=True, label=species_name)
    plt.title(f"Distribution of {feature} by Species")
    plt.xlabel(feature)
    plt.ylabel("Value")
    plt.legend(title="Species")
    plt.show()
```

```python
species_counts = iris_df["species"].value_counts()
```

```python
species_counts
```

```python
sns.barplot(x=species_counts.index, y=species_counts.values, palette="Set2")
plt.title("Species Count")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```

```python
correlation_matrix = iris_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Feature Correlation")
plt.show()
```

## Part 2 : Handling Null Values in Python with Pandas

```python
data = pd.read_csv("california_housing.csv")
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.dropna().shape
```

```python
data["AveOccup"].dropna().shape
```

```python
data["AveOccup"].isnull().sum()
```

```python
data = data.dropna(subset = ["AveOccup"])
```

```python
data.isnull().sum()
```

```python
# "total_bedrooms" column is not in the dataset so i'm using "AveBedrms" column for filling the Null Values
```

```python
data2 = data[["AveBedrms"]]
```

```python
data2
```

```python
data2.isnull().sum()
```

```python
#Replace the missing values in the "AveBedrms" column with the median value of the column.
data2["median_operation"] = data2["AveBedrms"].fillna(data2["AveBedrms"].median())
```

```python
data2
```

```python
#Fill the missing values in the "AveBedrms" column using interpolation.
data2["interpolation_operation"] = data2["AveBedrms"].interpolate(method = "linear")
```

```python
data2
```

```python
data2.describe()
```

```python
interpolated_data = data2[["interpolation_operation"]]
```

```python
interpolated_data
```

```python
plt.plot(data2["interpolation_operation"], marker = "o", color = "r", label = "interpolation_operation")
plt.xlabel("Index")
plt.ylabel("Interpolated Values")
plt.title("Line Plot of AveBedrms with Interpolated Values")
plt.legend()
plt.show()
```

```python
sns.lineplot(data = interpolated_data, marker = "o")
plt.xlabel("Index")
plt.ylabel("Interpolated Values")
plt.title("Line Plot of AveBedrms with Interpolated Values")
plt.show()
```
